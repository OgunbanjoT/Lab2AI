```python
# Task 1: Create a DataFrame using Pandas
# Create a DataFrame with the following data:
# Name: Alice, Bob, Charlie, David
# Age: 25, 30, 35, 28
# City: New York, San Francisco, Los Angeles, Chicago

# Your code here:
import pandas as pd
import numpy as np

# Dictionary containing the required data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
}

# Creating the DataFrame and ensuring correct data types
df = pd.DataFrame(data)
df['Age'] = df['Age'].astype(int)  # Explicitly cast to integer

print("Task 1 Output:")
print(df)
print("\nData Types:")
print(df.dtypes)
```

```python
# Task 2: Row and Column Manipulation
# Drop the 'City' column from the DataFrame created in Task 1

# Your code here:
try:
    # Dropping 'City' column using inplace=False to safeguard original data layout if needed
    df_dropped = df.drop(columns=['City'])

    print("Task 2 Output (City column dropped):")
    print(df_dropped)
    print("\nRemaining Columns:", df_dropped.columns.tolist())
except KeyError as e:
    print(f"Error: Column not found. {e}")
```

```python
# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

# Your code here:
# Creating a DataFrame with intentional NaN values across numeric and categorical columns
null_data = {
    'Product': ['Laptop', np.nan, 'Tablet', 'Monitor'],
    'Price': [1200, 850, np.nan, 300],
    'Quantity': [5, np.nan, 12, np.nan]
}
df_null = pd.DataFrame(null_data)
print("Original DataFrame with Null Values:")
print(df_null)

# Method 1: Fill specific columns with custom values (0 for numeric, 'Unknown' for text)
df_filled = df_null.copy()
df_filled['Product'] = df_filled['Product'].fillna('Unknown')
df_filled['Price'] = df_filled['Price'].fillna(0)
# Method 2: Backfill or forward fill alternative strategy for missing values
df_filled['Quantity'] = df_filled['Quantity'].ffill().fillna(0) # Forward fill, fallback to 0

print("\nTask 3 Output (Filled DataFrame):")
print(df_filled)
```

```python
# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:
# 1. Concatenate df1 and df2 vertically (axis=0) and reset index to avoid duplicate indices
df_concat = pd.concat([df1, df2], axis=0).reset_index(drop=True)

# 2. Merge/Join the concatenated result with df3 horizontally (axis=1)
# Note: df3 has 2 rows and df_concat has 4 rows; missing rows will naturally result in NaN
df_final = pd.concat([df_concat, df3], axis=1)

print("Task 5 Output (Combined DataFrame):")
print(df_final)
print(f"\nFinal Shape: {df_final.shape}")
```

```python
# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:
# Standard function definition
def square_num(x):
    """Calculates the square of a given numerical input."""
    try:
        return x ** 2
    except TypeError:
        return "Invalid input: Must be a number"

# Equivalent lambda function
square_lambda = lambda x: x ** 2 if isinstance(x, (int, float)) else "Invalid input"

# Testing inputs
test_inputs = [4, 7.5]

print("Task 8 Output:")
for val in test_inputs:
    print(f"Input: {val} -> Standard Function: {square_num(val)} | Lambda Function: {square_lambda(val)}")
```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# Class-based custom Iterator
class FirstFiveEvens:
    def __init__(self):
        self.count = 0
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= 5:
            raise StopIteration
        val = self.current
        self.current += 2
        self.count += 1
        return val

# Generator Function
def even_generator():
    count = 0
    current = 0
    while count < 5:
        yield current
        current += 2
        count += 1

# Demonstrate both
print("Iterator Results:")
iterator_obj = FirstFiveEvens()
print(list(iterator_obj))

print("\nGenerator Results:")
gen_obj = even_generator()
print(list(gen_obj))
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Your code here:
# 1. Map to square numbers
squared_numbers = list(map(lambda x: x**2, numbers))

# 2. Reduce to find the product
product_of_numbers = reduce(lambda x, y: x * y, numbers)

# 3. Filter to get even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print("Task 10 Output:")
print("Original List:", numbers)
print("Squared List (Map):", squared_numbers)
print("Product of List (Reduce):", product_of_numbers)
print("Even Numbers (Filter):", even_numbers)
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
class Rectangle:
    def __init__(self, length, width):
        # Validation to prevent invalid rectangle dimensions
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive values.")
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

# Instantiate and demonstrate two objects
try:
    rect1 = Rectangle(5, 10)
    rect2 = Rectangle(3.5, 4.2)

    print("Task 11 Output:")
    print(f"Rectangle 1 (5x10) -> Area: {rect1.calculate_area()}, Perimeter: {rect1.calculate_perimeter()}")
    print(f"Rectangle 2 (3.5x4.2) -> Area: {rect2.calculate_area():.2f}, Perimeter: {rect2.calculate_perimeter():.2f}")
except ValueError as e:
    print(f"Error creating object: {e}")
```

```python
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
# 1. Average salary by department
avg_salary_dept = df_employees.groupby('Department')['Salary'].mean()
print("1. Average Salary by Department:")
print(avg_salary_dept)
print("-" * 40)

# 2. Names of employees with salary > 60000
high_earners = df_employees[df_employees['Salary'] > 60000]['Name'].tolist()
print("2. Employees with Salary > 60,000:")
print(high_earners)
print("-" * 40)

# 3. Add a new column 'Bonus' (10% of salary)
df_employees['Bonus'] = df_employees['Salary'] * 0.10
print("3. Final DataFrame with Bonus Column:")
print(df_employees)
```

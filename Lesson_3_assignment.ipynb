```python
# Python Library Management System

# Install required libraries (This should already exist in colab)
!pip install requests

import pickle
import sqlite3
import requests

# 1. Data Types and Variables
book_title = "Python Programming"
book_quantity = 5
book_price = 29.99
is_available = True

print(f"Book: {book_title}, Quantity: {book_quantity}, Price: ${book_price}, Available: {is_available}")
```

```python
# 2. Lists
books = ["Python Basics", "Advanced Python", "Data Science with Python"]

# Add two more books to the list using append and insert
books.append("Deep Learning Foundations")
books.insert(1, "Machine Learning Guide")

print("Updated book list:", books)
```

```python
# 3. Loops and range()
print("\nAvailable books:")
# Use a for loop and range() to print books with their indices
for i in range(len(books)):
    print(f"Index {i}: {books[i]}")
```

```python
# 4. User Input
new_book = input("Enter a new book title: ")
books.append(new_book)
print("Updated book list:", books)
```

```python
# 5. Functions and Conditional Statements
def check_availability(book_name):
    # Case-insensitive check to handle edge cases gracefully
    cleaned_books = [book.lower() for book in books]

    if book_name.lower() in cleaned_books:
        return True
    else:
        return False

# Test the function
print(f"Is 'Python Basics' available? {check_availability('Python Basics')}")
print(f"Is 'Nonexistent Book' available? {check_availability('Nonexistent Book')}")
```

```python
# 6. File I/O
def save_books_to_file(filename):
    with open(filename, 'w') as f:
        for book in books:
            f.write(f"{book}\\n")

# Implement a function to read books from a file with error handling
def read_books_from_file(filename):
    try:
        with open(filename, 'r') as f:
            # .strip() removes trailing newline characters (\n)
            return [line.strip() for line in f.readlines()]
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return []

# Test file I/O
save_books_to_file("books.txt")
print("\n--- Displaying text file contents via system command ---")
!cat books.txt  # Display file contents in Colab

print("\nBooks verified via read_books_from_file function:")
print(read_books_from_file("books.txt"))
```

```python
# 7. Pickle for serialization
def save_books_pickle(filename):
    with open(filename, 'wb') as f:
        pickle.dump(books, f)

# Implement a function to load books using pickle with safety checks
def load_books_pickle(filename):
    try:
        with open(filename, 'rb') as f:
            return pickle.load(f)
    except FileNotFoundError:
        print(f"Error: The pickle file '{filename}' was not found.")
        return []
    except pickle.UnpicklingError:
        print("Error: Corrupted file data. Failed to parse pickle file.")
        return []

# Test pickle functions
save_books_pickle("books.pkl")
print("Books loaded from pickle:", load_books_pickle("books.pkl"))
```

```python
# 8. SQLite Database Interaction
def create_books_table():
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS books
                      (id INTEGER PRIMARY KEY, title TEXT, quantity INTEGER)''')
    conn.commit()
    conn.close()

# Implement a function to add a book using secure parameterized queries
def add_book_to_db(title, quantity):
    conn = None
    try:
        conn = sqlite3.connect('library.db')
        cursor = conn.cursor()

        # '?' placeholders protect against SQL injection and formatting breaks
        cursor.execute("INSERT INTO books (title, quantity) VALUES (?, ?)", (title, quantity))
        conn.commit()
        print(f"Successfully added '{title}' (Qty: {quantity}) to the database.")
    except sqlite3.Error as e:
        print(f"Database error while adding book: {e}")
    finally:
        if conn:
            conn.close()

# Test database functions
create_books_table()
add_book_to_db("Python for Beginners", 10)
```

```python
# 9. Web API Interaction (GitHub API)
def get_python_repos():
    url = "https://api.github.com/search/repositories"
    params = {"q": "language:python", "sort": "stars", "order": "desc"}
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  # Check for HTTP errors
        data = response.json()
        return [repo['name'] for repo in data['items'][:5]]
    except requests.RequestException as e:
        print(f"API Connection error: {e}")
        return []

# Test API function
print("Top 5 Python repositories on GitHub:", get_python_repos())

print("\nLibrary Management System operations completed.")
```
